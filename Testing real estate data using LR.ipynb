```python
X = data1.drop('Y house price of unit area', axis=1)
y = data1['Y house price of unit area']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.10, shuffle=False) 
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x_train, y_train) # train the model to set its parameter values and thereby make the model1 "trained"
```

```python
r_square = model.score(x_train, y_train) # coefficient of determination R^2 indicates correlation between features and target
print('coefficient of determination:', r_square) 
print('intercept:', model.intercept_)
print('coefficients:', model.coef_)
```

```python
y_pred = model.predict(x_test) # test step in sklearn
y_pred.reshape((-1,1))
```
